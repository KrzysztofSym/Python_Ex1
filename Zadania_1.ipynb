```python
# Stałe
podatek_dochod = 0.17

skl_emeryt = 0.0976
skl_rentow = 0.0150
skl_chorob = 0.0245

ubezp_zdrow_staw1 = 0.09
ubezp_zdrow_staw2 = 0.0775

koszt_uzyskania = 250.00
kwota_wolna = 43.76

placa_brutto = 3200.00

# Ubezpieczenia społeczne
emeryt = skl_emeryt * placa_brutto
rentow = skl_rentow * placa_brutto
chorob = skl_chorob * placa_brutto

spoleczn = emeryt + rentow + chorob

# Przychód
przychod = placa_brutto - spoleczn

# Ubezpieczenie zdrowotne
zdrow = ubezp_zdrow_staw1 * przychod

# Podstawa opodatkowania
podst_opod = przychod - koszt_uzyskania

# Podatek
podatek = podatek_dochod * podst_opod - kwota_wolna
podatek -= ubezp_zdrow_staw2 * przychod

# Płaca netto
placa_netto = placa_brutto - spoleczn - zdrow - podatek

# Wynik
print(
    "Płaca brutto: %f\n"
    "* Ub. społeczne: %f\n"
    "*** Ub. emeryt.: %f\n"
    "*** Ub. rentow.: %f\n"
    "*** Ub. chorob.: %f\n"
    "* Ub. zdrowotne: %f\n"
    "* Podatek doch.: %f\n"
    "Płaca netto: %f"
    % (
        placa_brutto,
        spoleczn,
        emeryt,
        rentow,
        chorob,
        zdrow,
        podatek,
        placa_netto
    )
)
```
